```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Cargar el conjunto de datos Fashion MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalizar los datos
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Definir el modelo de la red neuronal
model = keras.Sequential()

# Capa bidireccional de LSTM
model.add(layers.Bidirectional(layers.LSTM(256, return_sequences=True), input_shape=(28, 28))) # Asume imágenes de 28x28 píxeles
model.add(layers.Bidirectional(layers.LSTM(128)))

# Capas completamente conectadas
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.3))

# Capa de salida
model.add(layers.Dense(10, activation='softmax'))  # Capa de salida con 10 clases para Fashion MNIST

# Compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenar el modelo
model.fit(x_train, y_train, epochs=20, batch_size=64, validation_split=0.2)

# Evaluar el modelo en el conjunto de prueba
test_loss, test_acc = model.evaluate(x_test, y_test)
print('\nExactitud en el conjunto de prueba:', test_acc)
```

```python
model.summary()

import numpy as np
from sklearn.metrics import classification_report

# Suponiendo que ya tienes tu modelo entrenado y tus datos de prueba
# model es tu modelo entrenado y x_test son tus datos de prueba

# Realiza las predicciones en los datos de prueba
y_pred = model.predict(x_test)

# Convierte las predicciones de probabilidad en etiquetas de clase
y_pred_labels = np.argmax(y_pred, axis=1)

# Imprime el informe de clasificación
print(classification_report(y_test, y_pred_labels))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Cargar el conjunto de datos Fashion MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalizar los datos
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Definir el modelo de la red neuronal
model = keras.Sequential()

# Capa bidireccional de LSTM
model.add(layers.Bidirectional(layers.LSTM(256, return_sequences=True), input_shape=(28, 28))) # Asume imágenes de 28x28 píxeles
model.add(layers.Bidirectional(layers.LSTM(128)))

# Capas completamente conectadas
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.3))

# Capa de salida
model.add(layers.Dense(10, activation='softmax'))  # Capa de salida con 10 clases para Fashion MNIST

custom_optimizer = keras.optimizers.RMSprop(learning_rate=0.01)

# Compilar el modelo
model.compile(custom_optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenar el modelo
model.fit(x_train, y_train, epochs=20, batch_size=64, validation_split=0.2)

# Evaluar el modelo en el conjunto de prueba
test_loss, test_acc = model.evaluate(x_test, y_test)
print('\nExactitud en el conjunto de prueba:', test_acc)

model.summary()
```
